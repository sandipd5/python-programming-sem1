```python
def preprocess_text(text):
    
    text = ''.join(text)
    
    # Convert to lowercase
    text = text.lower()
    
    #Remove punctuation and special characters
    #re.sub(pattern, replacement, string)
    text = re.sub(r'[^\w\s]', '', text)
    
    # Tokenization
    tokens = word_tokenize(text)
    
    #Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token, wordnet.VERB) for token in tokens]

    # Remove repeated words and separate phrases
    tokens = [tokens[i] for i in range(len(tokens)) if i == 0 or tokens[i] != tokens[i - 1]]
    #join the tokens back to single stirng
    text = ' '.join(tokens)  
    
    return text
```

```python
df_intents['intents'] = df_intents['intents'].apply(preprocess_text)
```
